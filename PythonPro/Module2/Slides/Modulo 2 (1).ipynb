Vamos calcular o tempo de execução de um programa usando estrutura de repetição da maneira comum e utilizando list comprehension. Para isso, vamos criar um programa que printa na tela elementos de 0 a 10000 multiplicados por 2 e os adiciona em uma nova lista:

```python
import time

inicio=time.time()

for x in range(10001):
  print(x*2)

fim=time.time()

print(f'tempo de execução: {fim-inicio}')
```

Agora vamos criar uma matriz com notas de 3 alunos em 4 provas 3. Vamos também percorrer os valores, multiplicar por 2 e adicionar os novos elementos em um nova matriz. Veja:


```python
matriz=[[25.0,20.0,19.3,15.9],[18.7, 24.5, 17.8, 19.9], [22.4, 19.2, 15.9, 23.8]]
matriz
```

```python
import time
novaMatriz=[]


inicio=time.time()
for linha in matriz:
  listaNotas=[]
  for elemento in linha:
    listaNotas.append(elemento*2)
  novaMatriz.append(listaNotas)

fim=time.time()

print(f'tempo de execução: {fim-inicio}')


novaMatriz
```

```python
import time

inicio=time.time()

novaMatriz=[[elemento*2 for elemento in linha]for linha in matriz]

fim=time.time()

print(f'tempo de execução: {fim-inicio}')

novaMatriz
```
