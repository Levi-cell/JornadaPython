Crie um programa que contenha duas funções chamadas soma1 e soma2. Estas funções devem receber como parâmetro dois números (a e b) e calcular a soma de todos os outros números que existem entre a e b. A diferença é que em soma1, deve-se somar os valores usando o range() e sum(), e em soma2 deve-se domar usando estrutura de repetição for. 

a)Calcule o tempo de execução de cada função 

b)Agora paralelize este programa usando múltiplas threads

c)calcule o tempo de execução do programa paralelizado


```python
import time

def soma1(a, b):
    numeros=range(a, b+1)
    print(sum(numeros))

def soma2(a, b):
    soma = 0
    for numero in range(a, b+1):
        soma += numero
    print(soma)

inicio=time.time()
soma1(1,500000000)


soma2(1,500000000)

fim=time.time()

print(f"Tempo de execução = {fim-inicio}")
```

```python
from ast import arg
from threading import Thread
import time

def soma1(a, b):
    numeros=range(a, b+1)
    print(sum(numeros))

def soma2(a, b):
    soma = 0
    for numero in range(a, b+1):
        soma += numero
    print(soma)

inicio=time.time()

t1 = Thread(target=soma1, args=[1,500000000])

t2 = Thread(target=soma2, args=[1,500000000])

t1.start()
t2.start()

t2.join()
fim=time.time()

print(f"Tempo de execução = {fim-inicio}")

```
