```python
#Importação das bibliotecas
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("DADOS ALUNOS.xlsx")
df
```

a)

```python
from collections import Counter
from numpy.core.fromnumeric import mean

media=mean(df['Nota Prova 1'])+mean(df['Nota Prova 2'])+mean(df['Nota Prova 3'])+mean(df['Nota Seminario'])

maiorMedidaDF=df[(df['Nota Prova 1'])+mean(df['Nota Prova 2'])+mean(df['Nota Prova 3'])+mean(df['Nota Seminario'])>media]

maiorMedidaDF['Aluno(a)'].count()
```

b)

```python
participativos=df[df['Participativo nas aulas']=='S']

mean(participativos['Nota Prova 1'])+mean(participativos['Nota Prova 2'])+mean(participativos['Nota Prova 3'])+mean(participativos['Nota Seminario'])
```
