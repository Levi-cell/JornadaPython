```python
import pandas as pd
import datetime

df = pd.read_csv("1-dadosgovbr---2014.csv",sep = ';', encoding="latin-1")

df.head(15)
```

```python
#fornece todas as infos sobre a tabela
df.info()
```

```python
#pega a coluna da data da abertura e hora da mesma

data_hora_abertura = df[['Data Abertura','Hora Abertura']]
print(data_hora_abertura)

```

```python
#pega a coluna da data da resposta e hora

data_hora_resposta = df[['Data Resposta', 'Hora Resposta']]
print(data_hora_resposta)
```

```python
#tempo de resposta em dias apos analisar as info acima
tempo_resposta = df.get(['Tempo Resposta'])
print(tempo_resposta)
```

```python
# min med max dos tempos de resposta
# tentei pegar as horas tambem mais acabei me complicando um pouco
# usei o .round pra arrendondar os dados pois 6.3 dias ficaria estranho

minimo_dias = df['Tempo Resposta'].min()
max_dias = df['Tempo Resposta'].max()
soma_dias = df['Tempo Resposta'].sum()
med_dias = soma_dias / 43987
med_dias = med_dias.round()
print(f'A reclamacao respondida mais demorada foi realizada no tempo de {max_dias} dias!')
print(f'A reclamacao respondida mais rapida foi realizada no tempo de {minimo} dias!')
print(f'A media de resposta das reclamacoe foi realizada no tempo de {med_dias} dias!')
```

```python
# Notas min max med
nota = df['Nota do Consumidor']
nota_null = df['Nota do Consumidor'].isnull().sum()
print(nota)
print('A nota minima foi:',nota.min())
print('A nota maxima foi:',nota.max())
print('A media das notas e:',nota.sum() / 43987)
print('Numero de reclamacoes respondidas que nao foram avaliadas:', nota_null)
```

```python
# Tempo de resposta e nota correlacao!

# Selecionando as colunas desejadas
df_selecionado = df.loc[:,['Tempo Resposta', 'Nota do Consumidor']]

# Imprimindo o novo DataFrame
print(df_selecionado)

"""Nao podemos correlacionar a nota do consumidor com a resposta propriamente dita, pois o mesmo depende de outros fatores
nao apenas o quao rapido foi a resposta como se o problema foi resolvido, a parte do costumer-service etc pois por exemplo
podemos ver notas 5 que levou 5 dias para resposta, como notas 3 onde o individuo obteve a resposta no mesmo dia o que nao 
indica uma correlacao muito forte entre tempo de resposta e nota"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Contar a frequência de reclamações por sexo
frequencia_sexo = df['Sexo'].value_counts()

# Criar o gráfico de barras
plt.bar(frequencia_sexo.index, frequencia_sexo.values)

# Adicionar rótulos e título ao gráfico
plt.xlabel('Sexo')
plt.ylabel('Frequência de Reclamações')
plt.title('Frequência de Reclamações por Sexo')

# Exibir o gráfico
plt.show()
```

```python
# Contar a frequência de reclamações por estado
frequencia_estado = df.UF.value_counts()

# Criar o gráfico de barras
plt.bar(frequencia_estado.index, frequencia_estado.values)

# Adicionar rótulos e título ao gráfico
plt.xlabel('Estado')
plt.ylabel('Frequência de Reclamações')
plt.title('Frequência de Reclamações por Estado')

# Exibir o gráfico
plt.show()
```

```python

# Calcular a frequência de reclamações respondidas e não respondidas
respondida = df[df['Respondida'] == 'S'].shape[0]
nao_respondida = df[df['Respondida'] == 'N'].shape[0]

# Criar listas com as categorias e os valores correspondentes
categorias = ['Respondida', 'Não Respondida']
valores = [respondida, nao_respondida]

# Criar o gráfico de barras
plt.bar(categorias, valores)

# Adicionar rótulos e título ao gráfico
plt.xlabel('Situação da Reclamação')
plt.ylabel('Frequência de Reclamações')
plt.title('Frequência de Reclamações Respondidas e Não Respondidas')

# Exibir o gráfico
plt.show()
```

```python
a = df['Nota do Consumidor'].isnull().sum()
b = df[df['Nota do Consumidor'] == 1].shape[0]
c = df[df['Nota do Consumidor'] == 2].shape[0]
d = df[df['Nota do Consumidor'] == 3].shape[0]
e = df[df['Nota do Consumidor'] == 4].shape[0]
f = df[df['Nota do Consumidor'] == 5].shape[0]

print(f' Nota 1: {b} \n Nota 2: {c} \n Nota 3: {d} \n Nota 4: {e} \n Nota 5: {f} \n Nao Avaliada: {a}')
```
