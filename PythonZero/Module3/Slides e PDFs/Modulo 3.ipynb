Implemente o diagrama de classes a seguir:

**Regra de negócio: Só é possível sacar o saldo em conta e o salário não pode ser alterado por um salário menor.**

```python
class Pessoa:
  def __init__(self,nome, cpf, rg, endereco, telefone):
      self.nome=nome
      self.cpf=cpf
      self.rg=rg
      self.endereco=endereco
      self.telefone=telefone

  def alteraDados(self,nome=None, cpf=None, rg=None, endereco=None, telefone=None):
      if nome is not None:
        self.nome=nome
      if cpf is not None:
        self.cpf=cpf
      if rg is not None:
        self.rg=rg
      if endereco is not None:
        self.endereco=endereco
      if telefone is not None:
        self.telefone=telefone


  def MostraDados(self):
    print(f' Nome: {self.nome}\n CPF: {self.cpf}\n RG: {self.rg}\n Endereço: {self.endereco}\n Telefone: {self.telefone}')


class Funcionario(Pessoa):

  def __init__(self,nome, cpf, rg, endereco, telefone,salario, cargo, dtAdmissao):
      super().__init__(nome, cpf, rg, endereco, telefone)
      self.salario=salario
      self.cargo=cargo
      self.dtAdmissao=dtAdmissao


  def alteraDados(self,nome=None, cpf=None, rg=None, endereco=None, telefone=None, salario=None, cargo=None, dtAdmissao=None):
      super().alteraDados(nome, cpf, rg, endereco, telefone)
      if salario is not None:
        if salario> self.salario:
          self.salario=salario
        else:
          print('Não é possível alterar o salário pois ele é menor que o salário atual.')
      if cargo is not None:
        self.cargo=cargo
      if dtAdmissao is not None:
        self.dtAdmissao=dtAdmissao

  def MostraDados(self):
    super().MostraDados()
    print(f' Salario: {self.salario}\n Cargo: {self.cargo}\n Data de Admissão: {self.dtAdmissao}\n')



class Cliente(Pessoa):
  def __init__(self,nome, cpf, rg, endereco, telefone,conta, saldo, limite):
      super().__init__(nome, cpf, rg, endereco, telefone)

      self.conta=conta
      self.saldo=saldo
      self.limite=limite

  def sacar(self,valor):

      if self.saldo < valor:
          print('Você não tem este valor em conta, seu saldo é de: R$ ', self.saldo)
      else:
          self.saldo-=valor

  def depositar(self,valor):
      self.saldo+=valor


  def alteraDados(self,nome=None, cpf=None, rg=None, endereco=None, telefone=None, conta=None, saldo=None, limite=None):
      super().alteraDados(nome, cpf, rg, endereco, telefone)
      if conta is not None:
        self.conta=conta
      if saldo is not None:
        self.saldo=saldo
      if limite is not None:
        self.limite=limite

  def MostraDados(self):
    super().MostraDados()
    print(f' Conta: {self.conta}\n Saldo: {self.saldo}\n Limite: {self.limite}\n')





```
