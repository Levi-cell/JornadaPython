Vamos explorar a utilização dos módulos que trabalham com data e hora no python realizando as seguintes práticas:




1) Mostre a data de hoje conforme exemplo:
 "segunda, 02 de janeiro de 1994"


```python
# importando o módulo datetime
from datetime import datetime
 
#definindo a lista com os dias da semana em português
dias_nomes= [
    "Segunda",
    "Terça",
    "Quarta",
    "Quinta",
    "Sexta",
    "Sábado",
    "Domingo",
]

meses_nome= [
    "Janeiro",
    "Fevereiro",
    "Março",
    "Abril",
    "Maio",
    "Junho",
    "Julho",
    "Agosto",
    "Setembro",
    "Outubro",
    "Novembro",
    "Dezembro"
]

# recuperando a data e hora de hoje
hoje=datetime.now()

# imprimindo no formato do exxemplo
dia=hoje.day
diaNome=dias_nomes[hoje.weekday()]
mes=meses_nome[hoje.month-1]
ano=hoje.year

print(f"{diaNome}, {dia} de {mes} de {ano} ")
```

2) Converter um objeto datetime para date

```python
# importando o módulo datetime
from datetime import datetime

# recuperando a data e hora de agora
hoje=datetime.now()


#convertendo datetime para date. Ou seja, pegando apenas data de um datetime
data_hoje = hoje.date()

print(data_hoje)
```

3) Converter um objeto datetime para time

```python
# importando o módulo datetime
from datetime import datetime

# recuperando a data e hora de agora
hoje=datetime.now()


#convertendo datetime para time. Ou seja, pegando apenas hora de um datetime
hora_hoje = hoje.time()

print(hora_hoje)
```

4) Observe que o horário do exercício anterior está com o fuso horário incorreto. Converta para horário de Brasília.

Observe também que o horário de Brasília tem 3 horas de diferença da hora apresentada. Isso vai facilitar a conversão.

```python
# Importando o módulo datetime
import datetime

hora_modificada=datetime.time(hour=int(hora_hoje.hour-3), minute=hora_hoje.minute, second=hora_hoje.second)

print(hora_modificada)
```
