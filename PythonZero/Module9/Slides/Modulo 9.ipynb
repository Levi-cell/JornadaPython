Aqui vai um exemplo de como criar a data 29/01/2012 utilizando o módulo date do Python.


```python

# Importando o módulo datetime
import datetime
# Importando o módulo time
import time
 
# criando um objeto chamado data, que corresponde a data 29/01/2012
dataExemplo = datetime.date(year=2012, month=1, day=29)

print(dataExemplo)
```

Selecionando o dia

```python
dataExemplo.day
```

Selecionando mês

```python
dataExemplo.month
```

Selecionando ano

```python
dataExemplo.year
```

Exibindo data conforme formato brasileiro:

```python
dia=dataExemplo.day
mes=dataExemplo.month
ano=dataExemplo.year

print(f"{dia}/{mes}/{ano}")
```

Dia de hoje:

```python
hoje = datetime.date.today()

print(hoje)
```

Assim como criamos objetos do tipo date (data), podemos criar objetos do tipo time (hora). Veja:

```python
# criando um objeto chamado hora, que corresponde a hora 12:55:01
hora_example = datetime.time(hour=12, minute=55, second=1)
print(hora_example)
```

Podemos também selecionar, separadamente, hora minuto e segundo:

```python
hora=hora_example.hour
minuto=hora_example.minute
segundo=hora_example.second

print(f"hora: {hora}")
print(f"minuto: {minuto}")
print(f"segundo: {segundo}")
```

É possível, também, recuperar data e hora atual em um único objeto:

```python
agora = datetime.datetime.now()
print(agora)
```
